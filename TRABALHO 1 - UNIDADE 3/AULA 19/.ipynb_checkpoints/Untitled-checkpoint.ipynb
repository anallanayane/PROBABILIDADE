## Distribuição de Poisson

## Graficos PMF

```python
# Importando as bibliotecas necessarias
import matplotlib.pyplot as plt
import math
import seaborn as sns
import numpy as np
from scipy import linspace
from scipy.stats import poisson

# configuraçao seaborn
sns.set()
sns.set_style("white")

outcome_counts = linspace(0,10,11)

# Criando o poisson pmf para cada valor de lambda
probs_lambda1  = poisson.pmf(outcome_counts,1)
probs_lambda4  = poisson.pmf(outcome_counts,4)
probs_lambda10 = poisson.pmf(outcome_counts,10)

# plotar os graficos
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(20,12))

ax1.bar(x, probs_lambda1, label = 'lambda = 1')
ax1.legend(loc=0)
ax2.bar(x, probs_lambda4, label = 'lambda = 4')
ax2.legend(loc=0)
ax3.bar(x, probs_lambda10, label = 'lambda = 10')
ax3.legend(loc=0)
ax4.set_visible(False)

fig.text(0.1, 0.9,'PMF', color='black', size=15)
plt.show()
```

## Graficos CDF

```python
# Criando o poisson cdf para cada valor de lambda
probs_lambda1  = poisson.cdf(outcome_counts,1)
probs_lambda4  = poisson.cdf(outcome_counts,4)
probs_lambda10 = poisson.cdf(outcome_counts,10)

# plotar os graficos
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(20,12))

ax1.bar(x, probs_lambda1, label = 'lambda = 1')
ax1.legend(loc=0)
ax2.bar(x, probs_lambda4, label = 'lambda = 4')
ax2.legend(loc=0)
ax3.bar(x, probs_lambda10, label = 'lambda = 10')
ax3.legend(loc=0)
ax4.set_visible(False)

fig.text(0.1, 0.9,'CDF', color='black', size=15)
plt.show()
```

## Distribuição Poisson - Aproximação

```python
from scipy.stats import binom

sns.set_style("white")

b = binom.pmf(outcome_counts, 20,0.05)
p = poisson.pmf(outcome_counts, 1)

red = plt.scatter(outcome_counts, b, color ='red', label='X~binomial(20,0.05)')
blue = plt.scatter(outcome_counts, p, color='blue', label='Y~poisson(20*0.05)')

plt.legend(handles=[red, blue])

plt.show()
```

## Distribuição Hipergeométrica - PMF

```python
from scipy.stats import hypergeom

[M, n, N] = [100,5,4]
rv = hypergeom(M, n, N)
x = np.arange(0, 20+1)
pmf_b = rv.pmf(x)

fig = plt.figure()

leg = plt.plot(x, pmf_b, 'bo', color='b',label='Y~hypergeom(k,100,5,4)')
plt.legend(handles=leg)

plt.show()


```

## Distribuição Hipergeométrica - CDF

```python
[M, n, N] = [100,5,4]
rv = hypergeom(M, n, N)
x = np.arange(0, 20+1)
pmf_b = rv.cdf(x)

fig = plt.figure()
leg = plt.plot(x, pmf_b, 'bo', color='b',label='Y~hypergeom(k,100,5,4)')
plt.legend(handles=leg)

plt.show()
```
